```python
# tensorflow의 Sequential, Dense 기능을 불러오기
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense

# numpy->np, tensorflow->tf
import numpy as np
import tensorflow as tf

# "seed" 난수생성 함수
np.random.seed(3)
tf.random.set_seed(3)

#data 불러오기
Data_set = np.loadtxt("../dataset/ThoraricSurgery.csv", delimiter=",")

# X=환자의 기록, Y=수술 결과
X = Data_set[:,0:17]
Y = Data_set[:,17]

# model
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) # 노드, 입력데이터형식, 활성화함수
model.add(Dense(1, activation='sigmoid')) # 노드, 활성화함수

# 모델 학습
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #손실함수, 최적화방법, 정확도
model.fit(X, Y, epochs=100, batch_size=10)
```
